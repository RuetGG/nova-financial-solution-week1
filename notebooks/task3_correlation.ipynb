```python
import sys
import os

sys.path.append(os.path.abspath("..")) 
import pandas as pd
import glob
from src.analysis import sentiment, correlation, data_prep
from src.data.loader_cleaner import load_data

news = load_data("../data/raw_analyst_ratings.csv", parse_dates=['date'])
news['date'] = pd.to_datetime(news['date'], errors='coerce')


stock_files = glob.glob("../data/clean/*_metrics.csv")



```

```python
for f in stock_files:
    ticker = f.split("/")[-1].split("_")[0].split('\\')[1]
    
    stock_df = load_data(f, parse_dates=['Date'])
    
    news_stock = data_prep.align_news_stock(stock_df, news, ticker) 
    
    news_stock = sentiment.compute_sentiment(news_stock)
    daily_sent = sentiment.aggregate_daily_sentiment(news_stock)
    
    stock_df = correlation.calculate_daily_return(stock_df)
    corr, df_merged = correlation.corr_senti_return(stock_df, daily_sent, ticker)
    
    if corr is not None:
        print(f"{ticker} correlation: {corr:.3f}")
    else:
        print(ticker, "Not enough data")
   
    
```
